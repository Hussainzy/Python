**Row wise Concatenation:**

```python
concatenated_df=pd.concat([file1_df,file2_df])
print('Shape of  Row Wise Concatenated Df',concatenated_df.shape)
concatenated_df.head()
```

**Column Wise Concatenation:**

**Merge:**

The merge function is used to combine DataFrames based on common columns (similar to SQL joins).

**1) Inner Merge:** Combines rows that have common values in both DataFrames.
    **pd.merge(df1, df2, on='common_column')**

**2)  Outer Merge:** Combines all rows from both DataFrames, filling in NaNs where there are no matches.

**pd.merge(df1, df2, on='common_column', how='outer')**

**3) Left Merge:** Includes all rows from the left DataFrame and the matched rows from the right DataFrame.

**pd.merge(df1, df2, on='common_column', how='left')**

**4) Right Merge:** Includes all rows from the right DataFrame and the matched rows from the left DataFrame.

**pd.merge(df1, df2, on='common_column', how='right')**


```python
merged_df=pd.merge(file1_df, file2_df, on='Country')
merged_df
```

```python
merged_df1=pd.merge(file1_df, file2_df,on='Country',how='outer')
merged_df1
```

```python
merged_df2=pd.merge(file1_df, file2_df,on='Country',how='left')
merged_df2
```

```python
merged_df3=pd.merge(file1_df, file2_df,on='Country',how='right')
merged_df3
```
